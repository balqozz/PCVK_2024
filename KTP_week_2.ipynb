```python
import cv2  # import the cv2 module
import numpy as np
from skimage import io
from google.colab.patches import cv2_imshow  # import cv2_imshow from google.colab.patches

# Define the URL of the image
url = "https://images-wixmp-ed30a86b8c4ca887773594c2.wixmp.com/i/6aeaa80c-a19d-44ef-9ba3-a415103e4115/dc2ebiz-7b0ed39b-8d2e-4551-88bc-7d1f92fd8de5.png"  # Example URL

# Read the image from the URL
image = io.imread(url)

# Convert image from BGR to RGB
image_2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_3 = cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB)

# Get image dimensions
tinggi, lebar = image.shape[:2]

# Define coordinates for the top-left and bottom-right corners of the box
x1, y1 = 100, 230  # Top-left corner coordinates
x2, y2 = 160, 720  # Bottom-right corner coordinates

# Fill the box with random colors
for x in range(x1, x2):
    for y in range(y1, y2):
        if 0 <= x < tinggi and 0 <= y < lebar:  # Ensure coordinates are within image bounds
            random_color = np.random.randint(0, 256, size=3)  # Generate a random color
            image_3[x, y] = random_color  # Assign the random color to the pixel

# Combine the two images horizontally
final_frame = cv2.hconcat([image_2, image_3])

# Display the result
cv2_imshow(final_frame)  # Use cv2_imshow instead of cv2.imshow
cv2.waitKey(0)
cv2.destroyAllWindows()
```
