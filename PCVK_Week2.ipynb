# Tugas 5

```python
# Import library yang dibutuhkan
import random

# Buat gambar hitam
black_img3 = np.zeros(shape=(512,512,3), dtype=np.int32)

# Tentukan jumlah kotak
num_kotak = 5

# Loop untuk membuat kotak
for i in range(num_kotak):
  # Generate posisi, ukuran, dan warna acak
  x1 = random.randint(0, 400)
  y1 = random.randint(0, 400)
  x2 = random.randint(x1 + 50, 512)
  y2 = random.randint(y1 + 50, 512)
  color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

  # Gambar kotak pada gambar hitam
  cv.rectangle(black_img3, pt1=(x1,y1), pt2=(x2,y2), color=color, thickness=-1)

# Tampilkan gambar
plt.imshow(black_img3)
```

```python
img = cv.imread('/content/drive/MyDrive/kitten01.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
blue_channel = img[:,:]
blue_channel[:,:,0:2] = 0
img = blue_channel.copy()
height = blue_channel.shape[0]
width = blue_channel.shape[1]
rect_width = 50
rect_height = 50
pt1 = (0, height - rect_height)
pt2 = (rect_width, height)
cv.rectangle(img, pt1, pt2, color=(255,0,0), thickness=-1)

font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(img,text='/content/drive/MyDrive/kitten01.jpg',org=(70,100),fontFace=font,fontScale=0.450,color=(0,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(img)
```
